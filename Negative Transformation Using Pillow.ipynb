import modules

```python
from PIL import Image
from PIL import ImageFilter
import IPython.display as pydis
```

```python

#@title Load input images

input_image = '/content/sample_images/img_1.jpg' #@param {type:"string"}

img = Image.open(input_image);
```

```python
# image resize function 
def resize(img_src, basewidth, new_name):
  img = Image.open(img_src)
  wpercent = (basewidth/float(img.size[0]))
  hsize = int((float(img.size[1])*float(wpercent)))
  img = img.resize((basewidth,hsize), Image.ANTIALIAS)
  img.save(new_name)
  return img.size
```

```python
for i in range(0, img.size[0]-1):

    for j in range(0, img.size[1]-1):

        # Get pixel value at (x,y) position of the image

        pixelColorVals = img.getpixel((i,j));

       

        # Invert color

        redPixel    = 255 - pixelColorVals[0]; # Negate red pixel

        greenPixel  = 255 - pixelColorVals[1]; # Negate green pixel

        bluePixel   = 255 - pixelColorVals[2]; # Negate blue pixel

                   

        # Modify the image with the inverted pixel values

        img.putpixel((i,j),(redPixel, greenPixel, bluePixel));
```

```python
# save output images
img.save("output.jpg")

# input image save
img_copy = Image.open(input_image)
img_copy.save("input.jpg")
```

### Display output and input images

```python
# input and output images show
print("Input Image: ")

# resize and display
resize("input.jpg", 300, "input-preview.jpg")
pydis.display(pydis.Image('input-preview.jpg'))
```

```python
print("Output Image: ")

# resize and display
resize("output.jpg", 300, "output-preview.jpg")
pydis.display(pydis.Image('output-preview.jpg'))
```
