# First method: Steps for negative transformation

1. Read an image
2. Get height and width of the image
3. Each pixel contains 3 channels. So, take a pixel value and collect 3 channels in 3 different variables.
4. Negate 3 pixels values from 255 and store them again in pixel used before.
5. Do it for all pixel values present in image.

```python
# Read an image
img = cv2.imread(input_image, 1)


img_bgr = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_bgr)
plt.show()

# Histogram plotting of the image
color = ('r', 'g', 'b')
  
for i, col in enumerate(color):
      
    histr = cv2.calcHist([img_bgr], 
                         [i], None,
                         [256], 
                         [0, 256])
      
    plt.plot(histr, color = col)
      
    # Limit X - axis to 256
    plt.xlim([0, 256])
      
plt.show()
```

```python
# get height and width of the image
height, width, _ = img_bgr.shape
  
for i in range(0, height - 1):
    for j in range(0, width - 1):
          
        # Get the pixel value
        pixel = img_bgr[i, j]
          
        # Negate each channel by 
        # subtracting it from 255
          
        # 1st index contains red pixel
        pixel[0] = 255 - pixel[0]
          
        # 2nd index contains green pixel
        pixel[1] = 255 - pixel[1]
          
        # 3rd index contains blue pixel
        pixel[2] = 255 - pixel[2]
          
        # Store new values in the pixel
        img_bgr[i, j] = pixel
  
# Display the negative transformed image
plt.imshow(img_bgr)
plt.show()
  
# Histogram plotting of the
# negative transformed image
color = ('r', 'g', 'b')
  
for i, col in enumerate(color):
      
    histr = cv2.calcHist([img_bgr], 
                         [i], None,
                         [256],
                         [0, 256])
      
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
      
plt.show()
```

```python

# save image
im = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

cv2.imwrite("output_1.jpg", im)
```

#2nd method: Steps for negative transformation

1. Read an image and store it in a variable.
2. Subtract the variable from 1 and store the value in another variable.
3. All done. You successfully done the negative transformation.

```python
# Read an image
img_bgr = cv2.imread(input_image, 1)
img_bgr = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_bgr)
plt.show()
  
# Histogram plotting of original image
color = ('r', 'g', 'b')
  
for i, col in enumerate(color):
      
    histr = cv2.calcHist([img_bgr],
                         [i], None,
                         [256],
                         [0, 256])
      
    plt.plot(histr, color = col)
      
    # Limit X - axis to 256
    plt.xlim([0, 256])
      
plt.show()
```

```python

# Negate the original image
img_neg = 1 - img_bgr
  
plt.imshow(img_neg)
plt.show()
  
# Histogram plotting of
# negative transformed image
color = ('r', 'g', 'b')
  
for i, col in enumerate(color):
      
    histr = cv2.calcHist([img_neg],
                         [i], None, 
                         [256],
                         [0, 256])
      
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
      
plt.show()
```

```python
im = cv2.cvtColor(img_neg, cv2.COLOR_BGR2RGB)
cv2.imwrite("output_2.jpg", im)
```

#3rd method: Steps for negative transformation

```python
# Lets read the image.
img = cv2.imread(input_image)

# Convert the image to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Lets read the image in grayscale also.
gray = cv2.imread(input_image, 0)
```

```python
# Lets generate the negative image.
colored_negative = abs(255-img)
gray_negative = abs(255-gray)
```

```python
# Lets plot the results.
imgs = [img, gray, colored_negative, gray_negative]
title = ['coloured', 'gray', 'coloured-negative', 'gray-negative']


plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title(title[1])
plt.imshow(imgs[1], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 3)
plt.title(title[2])
plt.imshow(imgs[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 4)
plt.title(title[3])
plt.imshow(imgs[3], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python

# save image
im_col = cv2.cvtColor(colored_negative, cv2.COLOR_BGR2RGB)
cv2.imwrite("output_colored_negative.jpg", im_col)

im_gra = cv2.cvtColor(gray_negative, cv2.COLOR_BGR2RGB)
cv2.imwrite("output_gray.jpg", im_gra)
```
